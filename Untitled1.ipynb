```python
import zipfile, os

# Nama file zip dari upload
zip_path = list(uploaded.keys())[0]

# Ekstrak ke folder "covid_data"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("covid_data")

# Cek isi folder hasil ekstrak
os.listdir("covid_data")
```

```python
import pandas as pd

# Baca CSV
df = pd.read_csv("covid_data/covid_19_indonesia_time_series_all.csv")

# Lihat 5 baris pertama
df.head()
```

```python
# Lihat nama semua kolom
df.columns
```

```python
import matplotlib.pyplot as plt

# Filter data hanya untuk Indonesia
df_indonesia = df[df['Location'] == 'Indonesia']

# Plot grafik
plt.figure(figsize=(12,6))
plt.plot(df_indonesia['Date'], df_indonesia['Total Cases'], label='Total Cases')
plt.plot(df_indonesia['Date'], df_indonesia['Total Deaths'], label='Total Deaths')
plt.plot(df_indonesia['Date'], df_indonesia['Total Recovered'], label='Total Recovered')

# Format grafik
plt.xlabel('Date')
plt.ylabel('Jumlah Kasus')
plt.title('Perkembangan COVID-19 di Indonesia')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# Plot Recovery Rate
plt.figure(figsize=(12,6))
plt.plot(df_indonesia['Date'], df_indonesia['Case Recovered Rate'], color='green', label='Recovery Rate')

plt.xlabel('Date')
plt.ylabel('Recovery Rate (%)')
plt.title('Recovery Rate COVID-19 di Indonesia')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
# pastikan sudah import pandas sebagai pd sebelumnya
df_indonesia = df[df['Location'] == 'Indonesia'].copy()
df_indonesia['Date'] = pd.to_datetime(df_indonesia['Date'])
df_indonesia = df_indonesia.sort_values('Date')

# pastikan kolom Case Recovered Rate numeric
df_indonesia['Case Recovered Rate'] = pd.to_numeric(df_indonesia['Case Recovered Rate'], errors='coerce')
df_indonesia[['Date','Total Cases','Total Deaths','Total Recovered','Case Recovered Rate']].head()

```

```python
# Hitung ulang Case Recovered Rate (%)
df_indonesia['Case Recovered Rate'] = (df_indonesia['Total Recovered'] / df_indonesia['Total Cases']) * 100

# Ganti NaN dengan 0
df_indonesia['Case Recovered Rate'] = df_indonesia['Case Recovered Rate'].fillna(0)

# Cek hasilnya
df_indonesia.head(10)
```

```python
import plotly.express as px

# Grafik Total Kasus di Indonesia
fig = px.line(df_indonesia,
              x='Date',
              y='Total Cases',
              title='Total Kasus COVID-19 di Indonesia',
              labels={'Total Cases': 'Total Kasus', 'Date': 'Tanggal'})

# Simpan ke HTML
fig.write_html("index.html")

# Download file HTML
from google.colab import files
files.download("index.html")
```
